**2.** The encode function below will be a very useful starting point in your codec proect. Read it very carefully and make sure you understand what is going on.

An STFT is computed. At each time frame, the values and indices corresponding to the top frequencies (by magnitude) are saved. As an improvement to the previous labs' codec, we are now saving the full complex amplitudes, rather than just saving magnitude and discarding the phase.

```
def encode(a):
    S = librosa.stft(a.astype('float_'))                  # take the STFT
    nBins, nFrames = np.shape(S)                          # get the number of freq bins and number of time frames
    nPeaks = 20                                           # how many peaks to save at each time frame
    values = np.complex64( np.zeros((nPeaks, nFrames) ) ) # create an empty complex array to store the nPeaks complex values for each frame
    indices = np.zeros((nPeaks, nFrames)).astype('int')   # create an empty int array to store the indices for the nPeaks freqs at each frame

    for n in range(nFrames):                              # for each time frame n
        sort_idx = np.argsort( np.abs(S[:,n]) )           # sort the freq indices according to magnitude
        peak_idxs = sort_idx[-nPeaks:len(sort_idx)]       # form a list of the indices for the top nPeaks (by descending magnitude) 
        values[:,n] = S[peak_idxs, n]                     # store the values (located at the peak indices in S) into our values array
        indices[:,n] = peak_idxs                          # store the peak indices in our indices array

    return [indices, values]                              #return a list containing the indices and the values
```
